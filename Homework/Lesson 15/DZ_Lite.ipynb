### Домашнее задание: Создание простого блога на Django

**Цель:** Закрепить навыки установки Django, создания первого проекта и приложения, настройки маршрутов (URL), представлений (views) и работы с шаблонами (templates).

---

### Задание:

**1. Установка и создание проекта:**
- Установите Django через pip.
- Создайте новый проект Django с именем **"MyBlog"**.
- Создайте приложение **"blog"** для управления записями блога.

**2. Настройка проекта:**
- Настройте структуру проекта, убедитесь, что сервер разработки запускается корректно.
- Подключите приложение `blog` к проекту, зарегистрировав его в файле `settings.py`.

**3. Модели для блога:**
- Создайте модель **Post** с полями:
  - `title` — заголовок статьи.
  - `content` — содержание статьи.
  - `created_at` — дата создания (с автоматической записью времени).

  Выполните миграции для создания таблиц в базе данных.

**4. Настройка маршрутов (URLs):**
- Настройте маршруты в `urls.py` для вашего приложения:
  - Создайте путь для отображения всех записей блога на главной странице.
  
**5. Представления (views):**
- Создайте представление для вывода всех записей блога на странице.
  - Используйте `ListView` или функцию для отображения списка записей.

**6. Шаблоны (templates):**
- Создайте HTML-шаблон для отображения записей блога. Шаблон должен отображать заголовок и дату каждой записи.

---

### Результат выполнения:

После выполнения домашнего задания вы должны иметь простой работающий блог, где на главной странице отображаются все созданные записи с заголовками и датой создания.
